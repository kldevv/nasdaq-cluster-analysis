<h3><b> Get tickers(symbols & sectors) from "liberatedstocktrader.com"

```python
df = pd.read_html(r'https://www.liberatedstocktrader.com/nasdaq-100-companies-list-sector-market-cap/')[1]
df.columns = df.iloc[0, :]
nasdaq_sectors = df.drop(0, axis=0).loc[:, ['Symbol', 'SECTOR']]
tickers = nasdaq_sectors['Symbol']
# nasdaq_sectors.to_csv(r'../../raw_sectors/raw_sectors.csv')
nasdaq_sectors.head()
```

<h3><b> Get price data with yahoo_fin package 

```python
from yahoo_fin.stock_info import get_data, tickers_sp500, tickers_nasdaq, tickers_other, get_quote_table

# nasdaq_ticker_list = tickers_nasdaq()
# sp500_ticker_list = tickers_sp500()

# all_tickers = set(nasdaq_ticker_list + sp500_ticker_list)

ticker_prices_lookup = dict()
for idx, ticker in enumerate(tickers):
    print("Processing: {}, {}".format(idx, ticker))
    try:
        ticker_prices_lookup[ticker] = get_data(ticker, start_date="2015-01-01", end_date="2021-11-01")
    except (AssertionError, TypeError, KeyError, NameError) as err:
        print("{}: {}".format(ticker, err))
```
