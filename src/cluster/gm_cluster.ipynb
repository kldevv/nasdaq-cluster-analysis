```python
from sklearn.mixture import GaussianMixture
import pandas as pd

features = pd.read_csv(
    r"../../data/clean_market_data/features.csv", index_col=0)
features.head()
```

```python
gm_y = GaussianMixture(
    n_components=14, random_state=0).fit_predict(features)

gm_cluster = features.reset_index()
gm_cluster.insert(1, "Cluster", gm_y)

count_map = dict(gm_cluster['Cluster'].value_counts())
print(count_map)

gm_cluster['Cluster Weight'] = gm_cluster['Cluster'].map(count_map)

WEIGHT_TRESHOLD = int(gm_cluster.shape[0] / len(count_map))

gm_cluster = gm_cluster[gm_cluster['Cluster Weight']
                                    > WEIGHT_TRESHOLD]
gm_cluster.drop('Cluster Weight', axis=1, inplace=True)

print(gm_cluster['Cluster'].value_counts())

gm_cluster.to_csv(
    r'../../data/cluster_results/gm_cluster_result.csv')
```
