```python
import pandas as pd
```

```python
one_cluster = pd.read_csv(
    r"../../data/cluster_results/one_cluster_result.csv", index_col=0)
one_cluster.head()
```

```python
one_cluster.shape
```

```python
one_cluster_std = one_cluster.groupby("Cluster").std()
one_cluster_std
```

```python
one_cluster_std.mean(axis=1)
```

```python
one_cluster_std.median(axis=1)
```

```python
one_cluster_std.std(axis=1)
```
